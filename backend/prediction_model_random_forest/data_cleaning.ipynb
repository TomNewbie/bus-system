```python
import pandas as pd
import os
import random
import numpy as np
```

```python
FILE_PATH = os.path.join("dataset", "data.csv")
df = pd.read_csv(FILE_PATH, low_memory=False)
df
```

```python
df.columns
```

```python
# -- Choose features --
col = ['route_id', 'window', 'speed_kmh', 'avg_route_speed_kmh',
       'segment_max_speed_kmh', 'runtime_sec', 'segment_id', 'start_stop_id',
       'end_stop_id', 'distance_m', 'stop_lat', 'stop_lon', 'next_lat',
       'next_lon', 'arrival_time', 'departure_time']
df = df[col]
```

```python
def fix_time_format(time_str):
    try:
        # Split the time into hours, minutes, and seconds
        hours, minutes, seconds = map(int, time_str.split(':'))
        
        # Check if hours are greater than 23
        if hours > 23:
            # Correct the hours
            hours = 23
        
        # Reconstruct the corrected time string
        corrected_time = f'{hours:02d}:{minutes:02d}:{seconds:02d}'
        return corrected_time
    except:
        # Return the original value if it can't be corrected
        return time_str

# Apply the function to both columns
df['arrival_time'] = df['arrival_time'].apply(fix_time_format).copy()
df['departure_time'] = df['departure_time'].apply(fix_time_format).copy()
```

```python
df
```

```python
# -- Convert 'arrival-time' object to datetime64[ns] --
df['arrival_time'] = pd.to_datetime(df['arrival_time']).copy()
df['departure_time'] = pd.to_datetime(df['departure_time']).copy()
```

```python
# -- Create 'decisive_speed' attributes to determine the 'congestion_level' --
df['decisive_speed'] = df.apply(lambda row, weight=random.uniform(0.8, 0.9): (
        row['speed_kmh'] * weight + row['segment_max_speed_kmh'] * (1 - weight)), axis=1)
conditions = [
    (df['decisive_speed'] > 40),
    (df['decisive_speed'] > 30),
    (df['decisive_speed'] > 20),
    (df['decisive_speed'] > 15)
]
congestion_levels = [0, 1, 2, 3]
df['congestion_level'] = np.select(conditions, congestion_levels, default=4)
df = df.drop(['decisive_speed'], axis=1)
df = df.drop_duplicates(
    subset=['start_stop_id', 'arrival_time', 'stop_lat', 'stop_lon', 'route_id'
            , 'speed_kmh', 'segment_max_speed_kmh', 'runtime_sec', 'end_stop_id'])
```

```python
# -- Create 'arrival_hour' and 'arrival_minute' --
df = df[['route_id', 'window', 'speed_kmh', 'avg_route_speed_kmh',
       'segment_max_speed_kmh', 'runtime_sec', 'segment_id', 'start_stop_id',
       'end_stop_id', 'distance_m', 'stop_lat', 'stop_lon', 'next_lat',
       'next_lon', 'arrival_time', 'departure_time', 'congestion_level']]
```

```python
df['arrival_hour'] = df['arrival_time'].dt.hour.copy()
df['arrival_minute'] = df['arrival_time'].dt.minute.copy()

df = df[['route_id', 'window', 'speed_kmh', 'avg_route_speed_kmh',
       'segment_max_speed_kmh', 'runtime_sec', 'segment_id', 'start_stop_id',
       'end_stop_id', 'distance_m', 'stop_lat', 'stop_lon', 'next_lat',
       'next_lon', 'arrival_time', 'arrival_hour', 'arrival_minute', 'congestion_level']]
```

```python
df.to_csv('dataset\\clean_data.csv', index=False)
print(df.info())
```
