```python
import matplotlib.pyplot as plt


window_counts = df['window'].value_counts()

# Plot the bar chart
plt.bar(window_counts.index, window_counts.values)
plt.xlabel('Window')
plt.ylabel('Frequency')
plt.title('Distribution of Windows')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming your CSV file is named 'your_file.csv'
df = pd.read_csv('merged.csv')

# Assuming 'speed_kmh' is the target variable (you can change it based on your specific prediction goal)
target_variable = 'speed_kmh'

# Selecting relevant features for prediction
features = df[['stop_sequence', 'avg_route_speed_kmh', 'segment_max_speed_kmh']]

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, df[target_variable], test_size=0.2, random_state=42)

# Creating a linear regression model
model = LinearRegression()

# Training the model
model.fit(X_train, y_train)

# Making predictions on the test set
y_pred = model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Example: Predicting congestion for a new data point
new_data_point = np.array([[52, 1135.0, 1135.0]])  # Replace with actual values
prediction = model.predict(new_data_point)
print(f'Predicted Congestion Level: {prediction[0]}')
```
