```python
'''
    --- IMPORT DATA ---
'''

import pandas as pd


df = pd.read_csv("output.csv")

df.info()
```

```python
df.columns
```

```python
import numpy as np

conditions = [
    df['speed'].between(41, np.inf),
    df['speed'].between(31, 40),
    df['speed'].between(21, 30),
    df['speed'].between(16, 20),
    df['speed'].between(1, 15)
]
congestion_levels = [0, 1, 2, 3, 4]
df['congestion_level'] = np.select(conditions, congestion_levels, default=2)
```

```python
list(df.columns)
```

```python
df = df.dropna()
df
```

```python
df.to_csv("test.csv", index=False)
```

```python
part_1 = df.iloc[:125, :]
part_2 = df.iloc[125:250, :]
part_3 = df.iloc[250:375, :]
part_4 = df.iloc[375:, :]

# Save each part to a new CSV file
part_1.to_csv('part_1.csv', index=False)
part_2.to_csv('part_2.csv', index=False)
part_3.to_csv('part_3.csv', index=False)
part_4.to_csv('part_4.csv', index=False)
```
